```python
hello_data.plot(kind = "hist")
plt.title("Histogram of Diabetics Dataset")
plt.xlabel("Occurancies ")
plt.ylabel("Frequencies of Pregnacies")
plt.legend()
plt.show
```

```python

labels = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'
]
sizes = [
    3.845052, 120.894531, 69.105469, 20.536458,
    79.799479, 31.992578, 0.471876, 33.240885, 0.348958
]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Diabetic Dataset Attributes')
plt.show()
```

```python
hello_data.plot(kind = "line")
plt.xlabel("glucose")
plt.ylabel("BMI")
plt.title("Analyzig Diabetics dataset")
plt.show()
```

## conclusion for diabetic data 

###  This line chart appears to analyze relationships in a diabetes dataset, focusing on attributes like glucose, BMI, and others. Here are three possible conclusions:

Glucose and insulin levels exhibit significant variations compared to other variables, indicating their critical role in diabetes analysis.
BMI shows relatively consistent patterns, suggesting its stability as a factor across the dataset.
Other variables, like skin thickness and diabetes pedigree function, remain less variable but contribute to understanding diabetes' multifactorial nature.
